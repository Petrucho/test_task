1-я колонка – user_id – идентификатор пользователя. Длина user_id должна равняться 15-ти
символам. Идентификатор состоит из случайной комбинации следующих символов:
"1234567890abcdefghijk". Для каждой строки в DataFrame значение user_id формируются
случайным образом.

```python
import string
import random
import pandas as pd

userId_list = []
for x in range(10000):
    lst = [random.choice(string.ascii_letters + string.digits) for n in range(15)]    
    s = "".join(lst)
    userId_list.append(s)

df = pd.DataFrame(userId_list, columns=['user_id'])
```

```python
df.head(10)
print(f'\nlen(df): {len(df)}')
```

2-я колонка – order_number – номер заказа. Столбец необходимо заполнить случайными
значениями в диапазоне от 1 до 10. 

```python
orderId_list = []
for x in range(10000):
    lst = [random.choice(string.digits) for n in range(20)]    
    s = "".join(lst)
    orderId_list.append(s)

# print(f'orderId_list: {orderId_list}')
df['order_number'] = orderId_list
```

```python
df.head(10)
```

3-я колонка – click2delivery – время, прошедшее с момента оформления заказа до вручения
клиенту. Столбец необходимо заполнить случайными значениями из нормального распределения
со средним 1440 и стандартным отклонением 200.

```python
import numpy as np
import time

time_list = []
n = 10000

a=np.random.normal(loc=1440,scale=200,size=n)
# print(f'type(a): {type(a)}')
# print(f'len(a): {len(a)}')
# print(f'a: {a}')

for x in range(len(a)):
    time_to_string = time.strftime('%H:%M:%S', time.gmtime(a[x]))
    # print(f'time_to_string: {time_to_string}')
    time_list.append(time_to_string)

# print(f'\ntype(a): {type(a)}')
# print(f'\ntime_list: {time_list}')
df['click2delivery'] = time_list
```

4-я колонка – order_items_sum – общая стоимость заказа. Значения для этого столбца
необходимо взять из экспоненциального распределения с параметром λ = 1, смещённого на +1. 

```python
exp_list = []

for x in range(len(df)):
    exp_rand_plus_one = random.expovariate(lambd=1) + 1
    # print(f'exp_rand_plus_one: {exp_rand_plus_one}')
    exp_list.append(exp_rand_plus_one)

df['order_items_sum'] = exp_list
```

5-я колонка – retention – день жизни покупателя, в который он совершил заказ. Необходимо
сгенерировать значения 1, 2, 3, 4, 5 с вероятностями 0.35, 0.25, 0.2, 0.15 и 0.05 соответственно.

```python
retention_list = []

for x in range(10000):
    random_value = random.randint(1, 100) # 1..100
    if   random_value < 36: retention_list.append(1) # 35%
    elif random_value < 61: retention_list.append(2) # 25%
    elif random_value < 81: retention_list.append(3) # 20%
    elif random_value < 96: retention_list.append(4)# 15%
    else: retention_list.append(5) # 5%

# print(f'retention_list: {retention_list}')

df['retention'] = retention_list
```
